## 5. loc and iloc
Methods for accessing part of the DataFrame

### iloc (index location)
Uses number values for the indices of rows and columns

```python
# Get the element in the first row thrid column
df.iloc[0, 2]

# iloce slices
# Get the first 5 rows and first five columns
df.iloc[:5, :5]

# Get every row but only the second and third columns
df.iloc[ : , 1:3]

# Will give us the 6th row as a Series
df.iloc[5]

# Will give us the 6th row as a DataFrame
df.iloc[[5]]
```

### loc 
Uses index names and column names

```python
# Get every row but only certain columns
df.loc[:, ['BEDS', 'BATH']]

# Filters rows tieh loc and boolean expresions
df.loc[df['BEDS'] > 5]

# Filtering on multiple conditions(&, |, ~)(and, or, not)
df.loc[(df['BEDS'] > 5) & (df['BATH'] == 2)]
```

## 6. Querying Data

```python
# Basic Query
df.query("(BEDS > 5) and (BATH == 2)")

# YOu can use Strings in your queries by wrapping them in single quotes
df.query("SUBLOCALITY == 'Bronx County'")


df.query("SUBLOCALITY.str.contains('Bronx')")


county = "Bronx County"
df.query("SUBLOCALITY == @county")
```
