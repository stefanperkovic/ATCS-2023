# Simple Linear Regression
An example of how to create a Simple Linear Regression model using sklearn and data about crickets.

## 1. Import libraries and load data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("data/chirping.csv")
```

## 2. Determine if we can use Simple Linear Regression
* Data must be quantitative
* Data must have a linear relationship

```python
df.info()
```

```python
plt.title("Chirps vs Temp")
sns.scatterplot(data=df, x="Temp", y="Chirps")
```

## 3. Creating the Model

```python
# Extract our data
# X has to be a DataFrame while y is a Series
X_train = df[["Temp"]]
y_train = df["Chirps"]

# Create the model
model = LinearRegression()

# Train the model
model = model.fit(X_train, y_train)
```

## 4. Use the model for a prediction

```python
model.predict([[78], [95]])
```

## 5. Plot the Line of Best Fit

```python
plt.title("Chirps vs Temp")
sns.scatterplot(data=df, x="Temp", y="Chirps")
plt.plot(X_train, model.predict(X_train), color="r")
```

```python
df = pd.read_csv("data/advertising.csv")
df.info()
```

1. TV, radio, and newspaper are the features
Sales are the label

### Radio vs Sales

```python
# Extract our data
# X has to be a DataFrame while y is a Series
X_train = df[["radio"]]
y_train = df["sales"]

# Create the model
model = LinearRegression()

# Train the model
model = model.fit(X_train, y_train)
model.predict([[25]])
```

```python
model.predict([[65]])
```

```python
plt.title("Radio vs Sales")
sns.scatterplot(data=df, x="radio", y="sales")
plt.plot(X_train, model.predict(X_train), color="r")
```

Homework evaluating linear regression

```python
# Checking distribution of residuals
y_pred = model.predict(X_train)
residuals = y_train - y_pred
ax = sns.histplot(residuals)
```

Checking Homoscedasticity

```python
# Scatterplot residuals
sns.scatterplot(data=df, x="radio", y=residuals)
```

```python
model.score(X_train, y_train)
```

Looks pretty heterscedastic to me with being pretty grouped towards the begining displaying that it is better at predicting smaller values
